Matplotlib Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1)
y = x ** 2

plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', linestyle='--', label="y = x^2")

plt.title("Basic Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid()

plt.show()
```

SUB PLOTS

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, figsize=(8,6))

axs[0,0].plot(x, x)
axs[0,0].set_title("Linear")

axs[0,1].plot(x, x**2)
axs[0,1].set_title("Quadratic")

axs[1,0].plot(x, np.sin(x))
axs[1,0].set_title("Sine")

axs[1,1].plot(x, np.cos(x))
axs[1,1].set_title("Cosine")

plt.tight_layout()
plt.show()
```

Controlling Axes, Ticks, Labels

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)

plt.xlim(0,5)
plt.ylim(0,35)

plt.xticks([1,2,3,4], ['One','Two','Three','Four'], rotation=45)
plt.yticks([10,20,30], ['Low','Medium','High'])

plt.xlabel("Custom X Label", fontsize=12)
plt.ylabel("Custom Y Label", fontsize=12)

plt.title("Axis Control Example")

plt.show()
```

Legends

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

plt.plot(x, [10,20,25,30], label="Growth")
plt.plot(x, [30,25,20,15], label="Decline")

plt.legend(loc='upper right')
plt.title("Legend Example")

plt.show()
```

Annotation

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.plot(x, y, marker='o')

plt.annotate("Important Point",
             xy=(3,5),
             xytext=(4,8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.title("Annotation Example")
plt.grid()

plt.show()
```

Saving Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)

plt.title("Save Plot Example")

plt.savefig("output_plot.png", dpi=300)
plt.show()
```

Seaborn Scatter Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [2,5,3,8,10],
    "group": ["A","B","A","B","A"]
})

sns.scatterplot(data=data, x="x", y="y", hue="group")

plt.title("Seaborn Scatter Plot")
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

Text on Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [2,4,6]

plt.plot(x, y)

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})")

plt.title("Text Annotation")
plt.show()
```

3D Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1,2,3,4,5]
y = [2,4,3,5,1]
z = [10,15,20,25,30]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.title("3D Scatter Plot")
plt.show()
```

3D Line Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
z = [5, 3, 6, 2, 7]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot(x, y, z)

ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.title("3D Line Plot")

plt.show()
```

3D Surface Plot

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


surf = ax.plot_surface(X, Y, Z)

ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.title("3D Surface Plot")

plt.show()
```
